```
data = pd.read_csv('./btc.csv')
data.head()
```

```
# Clean number ins and outs
data['Number_ins'] = data['Number_ins'].astype(int)
data['Number_outs'] = data['Number_outs'].fillna(0).astype(int)
```

```
# Clean first in first out last in last out list time
def convert(x):
    if pd.notnull(x):
        return pd.to_datetime(x.split(' ')[0])
    else:
        return x

for col in ['First_in', 'Last_in', 'First_out', 'Last_out', 'List_time']:
    data[col] = data[col].apply(lambda x: convert(x))
```

```
data.head()
```

```
data.dropna(subset=['List_time'], inplace=True)
```

```
# Clean list amount
def convert(x):
    if x == '0 BTC':
        return 0
    x = x.split(' ')[0].replace(',', '')
    if x[0] == '-':
        return -float(x[1:])
    else:
        return float(x[1:])

data['List_amount'] = data['List_amount'].apply(lambda x: convert(x))
```

```
data.to_csv('./cleaned_data.csv', index=False)
```
